# **Control de Rendimiento en Extracción de Aceite**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento**:

Una planta procesadora de palma de aceite necesita calcular el **Índice de Rendimiento de Extracción (IRE)**. El operario debe ingresar el **peso total de los racimos de fruta fresca (RFF)** procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos**:

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo loat). (pr)
3. Solicitar los litros de aceite obtenidos (Tipo float). (lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)

$$pa = lt * 0.92$$

$$r = (pa/pr) * 100$$

**Prueba del programa**:

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.





```markdown
## Explicación del Código de Control de Rendimiento en Extracción de Aceite

El siguiente código Python realiza un cálculo del Índice de Rendimiento de Extracción (IRE) de aceite de palma, solicitando al usuario datos de entrada y mostrando el resultado.

```python
# ENTRADA Solicitar el nombre del lote de producción (Tipo string).
# Solicitar el peso de los racimos en kg (Tipo loat). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float). (lt)
```
Estas líneas son comentarios (`#`), que no son ejecutados por Python. Sirven para documentar el código, explicando qué se espera que hagan las siguientes líneas, en este caso, la **entrada** de datos y los tipos de variables esperados.

```python
nombre = input("Ingrese el nombre del lote de producción: ")
```
*   `input("Ingrese el nombre del lote de producción: ")`: Esta función muestra el mensaje entre comillas al usuario en la consola y espera a que el usuario ingrese un valor. El valor ingresado (que siempre es de tipo `string`) se asigna a la variable `nombre`.

```python
pr = float(input("Ingrese el peso de los racimos en kg: "))
```
*   `input("Ingrese el peso de los racimos en kg: ")`: Similar a la línea anterior, solicita al usuario el peso de los racimos.
*   `float(...)`: La entrada del usuario es un `string` por defecto. Esta función convierte ese `string` a un número de punto flotante (`float`), permitiendo valores decimales. El resultado se asigna a la variable `pr` (peso de los racimos).

```python
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))
```
*   `input("Ingrese la cantidad de aceite crudo obtenido en litros: ")`: Solicita al usuario la cantidad de aceite en litros.
*   `float(...)`: Convierte la entrada del usuario a un número de punto flotante. El resultado se asigna a la variable `lt` (litros de aceite).

```python
# PROCESO calcular peso del aceite (pa) y el rendimiento (r)
```
Este es otro comentario que indica que las siguientes líneas realizarán los cálculos (`PROCESO`).

```python
pa = lt * 0.92
```
*   `pa = ...`: Esta línea calcula el peso del aceite (`pa`).
*   `lt * 0.92`: Multiplica la cantidad de litros de aceite (`lt`) por su densidad (0.92 kg/L) para obtener el peso del aceite en kilogramos.

```python
r = (pa/pr) * 100
```
*   `r = ...`: Esta línea calcula el rendimiento (`r`).
*   `(pa/pr)`: Divide el peso del aceite (`pa`) por el peso total de los racimos (`pr`) para obtener la proporción de aceite.
*   `* 100`: Multiplica la proporción por 100 para expresarla como un porcentaje.

```python
# SALIDA Informar el rendimiento del lote
```
Este comentario indica que la siguiente línea mostrará el resultado (`SALIDA`).

```python
print("El rendimiento del lote", nombre, "es del", r, "%")
```
*   `print(...)`: Esta función muestra una salida al usuario en la consola.
*   Las comas `,` entre las cadenas de texto (`"..."`) y las variables (`nombre`, `r`) concatenan estos elementos en un solo mensaje. Python automáticamente añade un espacio entre cada elemento cuando se usa la coma en `print`. El mensaje final informará el nombre del lote y el rendimiento calculado en porcentaje.
```

````python
# ENTRADA Solicitar el nombre del lote de producción (Tipo string).
# Solicitar el peso de los racimos en kg (Tipo loat). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float). (lt)
nombre = input("Ingrese el nombre del lote de producción: ")
pr = float(input("Ingrese el peso de los racimos en kg: "))
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

# PROCESO calcular peso del aceite (pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa/pr) * 100

# SALIDA Informar el rendimiento del lote
print("El rendimiento del lote", nombre, "es del", r, "%")

````
