# EJERCICIO 1: CAPACIDAD DE SILOS DE GRANOS

## Planteamiento del Problema
Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico a partir del radio y la altura ingresados por el usuario.

## Requerimientos del problema
- Ingresar el radio (float)
- Ingresar la altura (float)
- Calcular el volumen del cilindro
- Multiplicar por la densidad del maíz

## Datos
Densidad = 0.75 ton/m³

Fórmula:
$V = \pi \cdot r^2 \cdot h$

## Prueba del programa
- Radio = 3 m  
- Altura = 10 m  
- Resultado esperado ≈ **212.06 toneladas**






```python
# ENTRADA Solicitar el radio y la altura del silo (Tipo float)
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# PROCESO Calcular el volumen del cilindro y la capacidad en toneladas
import math
volumen = math.pi * (radio ** 2) * altura
toneladas = volumen * 0.75

# SALIDA Mostrar la capacidad del silo
print("La capacidad del silo es:", round(toneladas, 2), "toneladas")
```

## Explicación del código

- **Línea 1:** Se solicita el radio del silo usando `input()` y se convierte a tipo `float`.
- **Línea 2:** Se solicita la altura del silo y también se convierte a `float`.
- **Línea 5:** Se importa la librería `math` para usar el valor de π.
- **Línea 6:** Se calcula el volumen del cilindro con la fórmula $V = \pi r^2 h$.
- **Línea 7:** Se multiplica el volumen por la densidad del maíz (0.75 ton/m³).
- **Línea 10:** Se imprime el resultado usando `print()` y se redondea con `round()`.
