# EJERCICIO 1: CAPACIDAD DE SILOS DE GRANOS

## 1.1. Planteamiento del Problema
Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico a partir del radio y la altura ingresados por el usuario.

## 1.2. Requerimientos del problema
- Ingresar el radio (float)
- Ingresar la altura (float)
- Calcular el volumen del cilindro
- Multiplicar por la densidad del maíz

## 1.3. Datos

Densidad = 0.75 ton/m³

Fórmula:
$$ V = \pi \cdot r^2 \cdot h $$

## 1.4. Prueba del programa
- Radio = 3 m  
- Altura = 10 m  
- Resultado esperado ≈ **212.06 toneladas**






```python
# ENTRADA Solicitar el radio y la altura del silo (Tipo float)
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# PROCESO Calcular el volumen del cilindro y la capacidad en toneladas
import math
volumen = math.pi * (radio ** 2) * altura
toneladas = volumen * 0.75

# SALIDA Mostrar la capacidad del silo
print("La capacidad del silo es:", round(toneladas, 2), "toneladas")
```

## 1.5. Explicación del código

- **Línea 1:** Se solicita el radio del silo usando `input()` y se convierte a tipo `float`.
- **Línea 2:** Se solicita la altura del silo usando `input()` y se convierte a tipo `float`.
- **Línea 3:** Comentario que indica la sección de entrada de datos.
- **Línea 4:** Se importa la librería math para utilizar funciones matemáticas como π.
- **Línea 5:** Se calcula el volumen del cilindro usando la fórmula $V = \pi r^2 h$.
- **Línea 6:** Se multiplica el volumen por 0.75 para convertirlo a toneladas.
- **Línea 7:** Comentario que indica la sección de salida de resultados.
- **Línea 8:** Se utiliza `print()` para mostrar el resultado y `round()` para redondear.

|----------|------|
| radio    | 3    |

## EJERCICIO 2: DOSIFICACIÓN DE FERTILIZANTE LÍQUIDO

## 2.1. Planteamiento del Problema
Se requiere calcular la cantidad de fertilizante y agua necesarios para un sistema de riego por goteo según las hectáreas y la dosis ingresadas.

## 2.2. Requerimientos del problema
- Ingresar las hectáreas (float)
- Ingresar la dosis en litros por hectárea (float)
- Calcular la cantidad total de fertilizante
- Calcular la cantidad de agua según la relación 1:500

## 2.3. Datos

Relación: 1 litro de fertilizante por 500 litros de agua

Fórmulas:

$$ Fertilizante = hectáreas * dosis $$  

$$ Agua = fertilizante * 500 $$

## 2.4. Prueba del programa
- Hectáreas = 5  
- Dosis = 2 L / ha  
- Resultado esperado: 10 L fertilizante y 5000 L agua




```python
# ENTRADA Solicitar hectáreas y dosis (Tipo float)
hectareas = float(input("Ingrese la cantidad de hectáreas: "))
dosis = float(input("Ingrese la dosis en L/ha: "))

# PROCESO Calcular fertilizante y agua
fertilizante = hectareas * dosis
agua = fertilizante * 500

# SALIDA Mostrar resultados
print("Se necesitan", int(fertilizante), "litros de fertilizante")
print("Se necesitan", int(agua), "litros de agua")
```

## 2.5. Explicación del código

- **Línea 1:** Se solicita la cantidad de hectáreas usando `input()` y se convierte a tipo `float`.
- **Línea 2:** Se solicita la dosis en litros por hectárea usando `input()` y se convierte a tipo float.
- **Línea 3:** Comentario que indica la sección de entrada de datos.
- **Línea 4:** Comentario que indica el inicio del proceso de cálculo.
- **Línea 5:** Se calcula la cantidad de fertilizante multiplicando hectáreas por dosis usando el operador *.
- **Línea 6:** Se calcula la cantidad de agua multiplicando el fertilizante por 500 usando el operador *.
- **Línea 7:** Comentario que indica la sección de salida de resultados.
- **Línea 8:** Se utiliza `print()` para mostrar los resultados y `int()` para eliminar los decimales.

## EJERCICIO 3: RENDIMIENTO DE PULPA EN FRUTAS

## 3.1. Planteamiento del Problema
Una planta procesadora necesita calcular el rendimiento de pulpa obtenida a partir de frutas, teniendo en cuenta el peso total y los residuos generados (cáscaras y semillas).

## 3.2. Requerimientos del problema
- Ingresar el peso total de la fruta (Tipo float)
- Ingresar el peso de los residuos (Tipo float)
- Calcular la cantidad de pulpa
- Calcular el porcentaje de rendimiento

## 3.3. Datos

Pulpa = peso total - residuos  

Rendimiento = (Pulpa / peso total) * 100

## 3.4. Prueba del programa
Peso total = 500 kg  
Residuos = 150 kg  
Resultado esperado: 350 kg de pulpa (70%)*texto en cursiva*

```python
# ENTRADA Solicitar peso total y residuos (Tipo float)
total = float(input("Ingrese el peso total de la fruta en kg: "))
residuos = float(input("Ingrese el peso de los residuos en kg: "))

# PROCESO Calcular pulpa y rendimiento
pulpa = total - residuos
rendimiento = (pulpa / total) * 100

# SALIDA Mostrar resultados
print("La cantidad de pulpa es:", int(pulpa), "kg")
print("El rendimiento es:", int(rendimiento), "%")
```

## 3.5. Explicación del código

- **Línea 1:** Se solicita el peso total de la fruta usando `input()` y se convierte a tipo float.
- **Línea 2:** Se solicita el peso de los residuos usando `input()` y se convierte a tipo float.
- **Línea 3:** Comentario que indica la sección de entrada de datos.
- **Línea 4:** Comentario que indica el inicio del proceso de cálculo.
- **Línea 5:** Se calcula la pulpa restando los residuos al peso total usando el operador -.
- **Línea 6:** Se calcula el rendimiento utilizando la fórmula `(pulpa / total) * 100`, usando los operadores / y  *.
- **Línea 7:** Comentario que indica la sección de salida de resultados.
- **Línea 8:** Se utiliza `print()` para mostrar los resultados y `int()` para eliminar los decimales.

## 3.6. Prueba de escritorio

| Variable     | Valor |
|-------------|------|
| total       | 500  |
| residuos    | 150  |
| pulpa       | 350  |
